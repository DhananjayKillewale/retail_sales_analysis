```python
# Step 1: Read the raw single column
raw = pd.read_excel(r'C:\Users\sunay\OneDrive\Dokumen\cleaned_retail_data.csv.xlsx', header=0)

# Step 2: Extract the column as text and split by comma
from io import StringIO

# Join all rows into a CSV-like string and re-parse
col_name = raw.columns[0]  # get the messy column header
csv_string = col_name + '\n' + '\n'.join(raw.iloc[:, 0].astype(str))

csv_data = pd.read_csv(StringIO(csv_string))

# Step 3: Verify
print("Column count:", len(csv_data.columns))
print("Columns:", csv_data.columns.tolist())
print("Shape:", csv_data.shape)
print(csv_data.head())
```

```python
csv_data['order_date'] = pd.to_datetime(csv_data['order_date'], errors='coerce')
```

```python
csv_data.loc[:, 'quantity'] = pd.to_numeric(csv_data['quantity'], errors='coerce')
```

```python
csv_data.loc[:, 'city'] = csv_data['city'].fillna('Unknown')
```

```python
csv_data.loc[:, 'product'] = csv_data['product'].fillna('Unknown')
```

```python
csv_data.loc[:, 'price'] = csv_data['price'].fillna(csv_data['price'].median())
```

```python
csv_data.loc[:, 'quantity'] = csv_data['price'].fillna(csv_data['quantity'].median())
```

```python
csv_data.loc[:,'category'] = csv_data['category'].str.strip().str.capitalize()
```

```python
print(csv_data.info())
```

```python
print(csv_data.head())
```

```python
csv_data.loc[:, 'order_time'] = pd.to_datetime(csv_data['order_time'], format='%H:%M:%S', errors='coerce')
```

```python
csv_data = csv_data.dropna(subset=['order_id', 'order_date'])
```

```python
csv_data.loc[:, 'quantity'] = csv_data['quantity'].fillna(csv_data['quantity'].median())
```

```python
csv_data.to_csv('cleaned_retail_data.csv', index=False)
```
